#### 1. Plot the data

```python
import matplotlib.pyplot as plt

# Create lists for the x and y coordinates of the triangle's vertices
x = [3, 8, 14, 3]
y = [4, 13, 5, 4]

# Plot the triangle using the 'plot' function
plt.plot(x, y, linestyle='-', marker='o', color='red')

# Add labels for the x and y axes
plt.xlabel("X")
plt.ylabel("Y")

# Show points and A, B, and C
point = ['A', 'B', 'C', 'A']
for i in range(len(x)):
    plt.text(x[i], y[i], f'{point[i]} ({x[i]}, {y[i]})')

# Add a title to the graph
plt.title("Distance between points A, B, and C")

# Show the graph
plt.show()
```

#### 2. Calculate the distance with the Euclidean distance formula

```python
from sklearn.metrics.pairwise import euclidean_distances

# Define the coordinates of the points
a = [[3, 4]]
b = [[8, 13]]
c = [[14, 5]]

# Calculate the Euclidean distance between the points
ab = euclidean_distances(a, b)
bc = euclidean_distances(b, c)
ca = euclidean_distances(c, a)

# print(ab)

# Print the distances
print(f"The distance between A and B is {ab[0][0]:.3f}")
print(f"The distance between B and C is {bc[0][0]:.3f}")
print(f"The distance between C and A is {ca[0][0]:.3f}")
```
